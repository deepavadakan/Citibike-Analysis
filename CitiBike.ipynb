```python
import pandas as pd
import glob
```

### Concatinating 2019 citibike csv files

```python
#bike_df = pd.read_csv("202012-citibike-tripdata.csv")
bike_2019_df = pd.concat(map(pd.read_csv, glob.glob('Resources/2019/*.csv')))
bike_2019_df
```

## Convert Date columns

```python
bike_2019_df['starttime']=pd.to_datetime(bike_2019_df['starttime'])
bike_2019_df['stoptime']=pd.to_datetime(bike_2019_df['stoptime'])
bike_2019_df
```

## Grouping bikes by month and start station

```python
start_stations_2019_df = bike_2019_df.groupby([bike_2019_df['starttime'].dt.month, 'start station id']).agg({'bikeid':'count',
                                                      'tripduration': 'mean',
                                                       'start station name':'first',
                                                       'start station latitude':'first',
                                                       'start station longitude':'first'})
start_stations_2019_df = start_stations_2019_df.reset_index(level=[0,1])

start_stations_2019_df = start_stations_2019_df.rename(columns={'starttime' : 'start month',
                                                    'start station id': 'station id',
                                                    'tripduration': 'avg trip duration', 
                                                    'bikeid':"number of bikes",
                                                    'start station name': 'station name',
                                                    'start station latitude': "latitude",
                                                    'start station longitude': 'longitude'})
start_stations_2019_df['start month']=pd.to_datetime(start_stations_2019_df['start month'],format="%m")
start_stations_2019_df['start month'] = start_stations_2019_df['start month'].dt.month_name()
#start_stations_2019_df
start_stations_2019_df.to_csv('data/2019_citibikes_start_stations.csv', index=False)
```

```python
end_stations_2019_df = bike_2019_df.groupby([bike_2019_df['stoptime'].dt.month, 'end station id']).agg({'bikeid':'count',
                                                      'tripduration': 'mean',
                                                       'end station name':'first',
                                                       'end station latitude':'first',
                                                       'end station longitude':'first'})
end_stations_2019_df = end_stations_2019_df.reset_index(level=[0,1])

end_stations_2019_df = end_stations_2019_df.rename(columns={'stoptime' : 'stop month',
                                                    'end station id': 'station id',
                                                    'tripduration': 'avg trip duration', 
                                                    'bikeid':"number of bikes",
                                                    'end station name': 'station name',
                                                    'end station latitude': "latitude",
                                                    'end station longitude': 'longitude'})
end_stations_2019_df['stop month']=pd.to_datetime(end_stations_2019_df['stop month'],format="%m")
end_stations_2019_df['stop month'] = end_stations_2019_df['stop month'].dt.month_name()
#start_stations_2019_df
end_stations_2019_df.to_csv('data/2019_citibikes_end_stations.csv', index=False)
```

## Group data by usertype

```python
bike_2019_usertype_df = bike_2019_df.groupby([bike_2019_df['starttime'].dt.month, "usertype"]).agg({'bikeid':'count',
                                                      'tripduration': 'mean'})
bike_2019_usertype_df = bike_2019_usertype_df.reset_index(level=[0,1])
bike_2019_usertype_df = bike_2019_usertype_df.rename(columns={'starttime' : 'month',
                                                    'tripduration': 'avg trip duration', 
                                                    'bikeid':"number of bikes"})
bike_2019_usertype_df['month']=pd.to_datetime(bike_2019_usertype_df['month'],format="%m")
bike_2019_usertype_df['month'] = bike_2019_usertype_df['month'].dt.month_name()
bike_2019_usertype_df
bike_2019_usertype_df.to_csv('data/2019_citibikes_usertype.csv', index=False)
```

## Group data by Age group and gender

```python
#create the bins for the age groups
age_bins = [10, 19, 29, 39, 49, 59, 69, 100]

#create the labels for the age groups
age_group_names = ["<20", "20-29", "30-39", "40-49", "50-59", "60-69", "70+"]

bike_2019_df['age'] = 2019 - bike_2019_df['birth year']
bike_2019_df
```

```python
#splice the data into the bins
age_series = pd.cut(bike_2019_df["age"], age_bins, labels=age_group_names, include_lowest=True)

#group the data by age groups
bike_2019_df['Age Ranges'] = age_series

bike_2019_age_ranges_df = bike_2019_df.groupby([bike_2019_df['starttime'].dt.month, bike_2019_df['Age Ranges'], "gender"]).agg({'bikeid':'count',
                                                       'tripduration': 'mean'})

bike_2019_age_ranges_df = bike_2019_age_ranges_df.reset_index(level=[0,1,2])
bike_2019_age_ranges_df = bike_2019_age_ranges_df.rename(columns={'starttime' : 'month',
                                                    'tripduration': 'avg trip duration', 
                                                    'bikeid':"number of bikes"})
bike_2019_age_ranges_df['month']=pd.to_datetime(bike_2019_age_ranges_df['month'],format="%m")
bike_2019_age_ranges_df['month'] = bike_2019_age_ranges_df['month'].dt.month_name()
bike_2019_age_ranges_df
bike_2019_age_ranges_df.to_csv('data/2019_citibikes_age_gender.csv', index=False)
```
